```r
complaint_delivery_analysis <- sqldf("

SELECT
    c.order_id,
    c.severity,
    d.delivery_status,
    d.route_distance_km,
    d.customer_rating_post_delivery

FROM complaints c

JOIN deliveries d
ON c.order_id = d.order_id

")

head(complaint_delivery_analysis)
```

```r
incident_complaint_analysis <- sqldf("

SELECT
    i.incident_id,
    i.severity,
    c.complaint_type,
    c.status,
    c.resolution_days

FROM incidents i

JOIN deliveries d
ON i.delivery_id = d.delivery_id

JOIN complaints c
ON d.order_id = c.order_id

LIMIT 10

")

head(incident_complaint_analysis)
```
